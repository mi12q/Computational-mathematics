```python
import numpy as np
from scipy import linalg
from scipy.linalg import norm
import matplotlib.pyplot as plt
import math
```

# Отделение корней графическим методом

```python
def graph():
    
    fig, ax = plt.subplots(figsize=(7, 5), dpi=150)
    theta = np.linspace(0, 2 * np.pi, 100)
    r = 1
    x1 = r * np.cos(theta)
    y1 = r * np.sin(theta)
    plt.axis('equal')
    ax.plot(x1,y1, color = "seagreen",label = 'x^2 + y^2 = 1')
    
    x2=np.arange(-1, 1, step = 0.1)
    y2 = []
    for i in x1:
        y2.append(math.tan(i))  
     
    ax.plot(x1,y2, color = "darkred",label = "y = tgx")
    plt.minorticks_on()
    plt.grid(which='major',
            color = 'grey', 
            linewidth = 0.7)
    plt.grid(which='minor', 
            color = 'grey', 
            linestyle = ':')
    ax.legend()
    ax.axhline(y=0, color='k')
    ax.axvline(x=0, color='k')
    plt.grid()

graph()
```

# Правило Нютона

```python
def newtons_method(eps,x0,y0):
    n = 0
    u0 = np.array([x0,y0])
    f0 = np.array([x0**2+y0**2-1,y0-math.tan(x0)])
    J0 = np.array([[2*x0,2*y0],[-1/math.cos(x0)**2,1]])
    u = u0 - (linalg.inv(J0)).dot(f0)
    print("Итерация {}:  x = {}, y = {} ".format(n,u[0],u[1]))
    while (norm(u-u0,1) > eps):
        n +=1
        u0 = u
        f = np.array([u[0]**2+u[1]**2-1,u[1]-math.tan(u[0])])
        J = np.array([[2*u[0],2*u[1]],[-1/math.cos(u[0])**2,1]])
        u = u0 - (linalg.inv(J)).dot(f)
        print("Итерация {}:  x = {}, y = {} ".format(n,u[0],u[1]))
    return u

solution1 = newtons_method(10**(-6),-1,-1)
print("Решение 1:  x = {}, y = {} ".format(solution1[0],solution1[1]))
print("\n")

solution2 = newtons_method(10**(-6),1,1)
print("Решение 2:  x = {}, y = {} ".format(solution2[0],solution2[1]))
```
