```python
import numpy as np
import matplotlib.pyplot as plt
import math

def f(x):
    f = math.atan(1/(x+10**(-8))) - math.pi/2 + 10**(-8)
    return math.atan(1/(x+10**(-8))) - math.pi/2 + 10**(-8)
    
def df(x):
    der = -1/(1+(x+10**(-8))**2)
    return der
    
def ddf(x):
    return (10000000000000000*(20000000000000000*x+200000000))/(10000000000000000*x**2+200000000*x+10000000000000001)**2
    
def graph():
    x=np.arange(-2*np.pi, 2*np.pi, step = 0.1)
    y = []
    dy = []
    ddy = []
    for i in x:
        y.append(f(i))
        dy.append(df(i))
        ddy.append(ddf(i))
    fig, ax = plt.subplots(figsize=(7, 5), dpi=150)
    ax.plot(x,y, color = "blue",label = 'y(x)')
    ax.plot(x,dy, color = "green",label = "y'(x)")
    ax.plot(x,ddy, color = "red",label = 'y"(x)')
    plt.xlim([-3.5, 3.5])
    plt.ylim([-3.5, 3.5])
    plt.minorticks_on()
    plt.grid(which='major',
            color = 'grey', 
            linewidth = 0.7)
    plt.grid(which='minor', 
            color = 'grey', 
            linestyle = ':')
    ax.legend()
    ax.axhline(y=0, color='k')
    ax.axvline(x=0, color='k')
    plt.grid()

graph()
```

```python
def newtons_method(eps):
    n = 0
    x0 = 1*10**(-3)
    print("f = ",f(x0))
    print("df = ",df(x0))
    print("d2f = ",ddf(x0))
    m = abs(df(x0))
    M = abs(ddf(x0))
    # m = 1
    # M = 2*10**(-8)
    h = f(x0)/df(x0)
    x = x0 - h
    print("f = ",f(x))
    print("df = ",df(x))
    print("d2f = ",ddf(x))
    # print("Итерация" + str(n),  "{:.4e}".format(x))
    # print(abs(x-x0)**2*M/(2*m))
    while abs(x-x0)**2*M/(2*m) > eps:  
        # print(abs(x-x0)**2*M/(2*m))
        n+=1
        x0 = x
        h = f(x) / df(x)
        x = x - h
        print("f = ",f(x))
        print("df = ",df(x))
        print("d2f = ",ddf(x))
        # print("Итерация" + str(n),  "{:.4e}".format(x))
    return x

x = newtons_method(10**(-10))
print("\n")
print("Решение: ",  "{:.4e}".format(x))
```
