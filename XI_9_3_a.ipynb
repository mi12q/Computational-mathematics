```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def iteration(y0,dy0,x,h):
    y = y0 + h*dy0zline = np.linspace(0, 15, 1000)
    dy = dy0 + h*(x*math.sqrt(y0))
    return y, dy
```

```python
def solve_(y0,dy0,x0,x1):
    y = []; x = x0
    h = 10**(-3)
    while x < x1:
        y0, dy0 = iteration(y0,dy0,x,h)
        y.append(y0)
        x += h
    return y0, y
```

```python
Y0 = 2
h = 10**(-3)
F = lambda alpha: solve_(0,alpha,0,1)[0] - Y0
dF = lambda alpha, h: (F(alpha + h)-F(alpha))/h
```

```python
alpha0 = 2
print(solve_(0,alpha0,0,1)[0])
f = F(alpha0)
print(f)
```

```python
alpha = alpha0
while(abs(f) > 10**(-12)):
    alpha = alpha - F(alpha)/dF(alpha,h)
    f = F(alpha)
```

```python
solution,y = solve_(0,alpha,0,1)
print(solution)
```

```python
x = np.linspace(0,1,len(y))
fig, ax = plt.subplots(figsize=(5, 4), dpi=150)
ax.plot(x,y, color = "darkred",label = "y(x)")
plt.minorticks_on()
plt.grid(which='major',
            color = 'grey', 
            linewidth = 0.7)
plt.grid(which='minor', 
            color = 'grey', 
            linestyle = ':')
plt.xlabel('x')
plt.ylabel('y')
ax.legend()
```
