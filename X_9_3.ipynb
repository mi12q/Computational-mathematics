```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def iteration(y0,dy0,mu,T,x0,dx0,h):
    y = y0 + h*dy0
    dy = dy0 + h*(mu*(1-dy0**2)*dy0 - y0)
    return y, dy
```

```python
def solve_Reley(mu,t0,y0,dy0):
    y = []; t = 0
    h = 10**(-3)
    while t < t0:
        y0, dy0 = iteration(y0,dy0,mu,t0,y0,dy0,h)
        y.append(y0)
        t += h
    return y
```

```python
y = solve_Reley(1000,1000,0,0.001)
x = np.linspace(0,1000,len(y))
fig, ax = plt.subplots(figsize=(5, 4), dpi=150)
ax.plot(x,y, color = "darkred",label = "y(t)")
plt.minorticks_on()
plt.grid(which='major',
            color = 'grey', 
            linewidth = 0.7)
plt.grid(which='minor', 
            color = 'grey', 
            linestyle = ':')
plt.xlabel('t')
plt.ylabel('y')
ax.legend()
```
