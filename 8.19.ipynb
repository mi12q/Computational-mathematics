```python
import numpy as np
import math as math
import matplotlib.pyplot as plt
```

# Вычисление нужного числа слагаемих для достижения данной точности

```python
def sin_series(x,n):
    S = 0
    for i in range(0,n):
        S += (((((-1)**i))*(x**((2*i)+1)))/(math.factorial((2*i)+1)))
    return S
    
def exp_series(x,n):
    S = 0
    for i in range(0,n):
        S += x**(i)/math.factorial(i)
    return S
    
def n_sin(b,eps):
    x = b
    n = 0
    s = (x**((2*(n+1)+1)))/math.factorial(2*(n+1)+1)
    while (s > eps):
        n += 1
        s = (x**((2*(n+1))+1))/math.factorial(2*(n+1)+1)
    return n

def n_exponent(b,eps):
    x = b
    n = 0
    s = x**(n+1)/math.factorial(n+1)
    while(s > eps):
        n += 1
        s = x**(n+1)/math.factorial(n+1)
    return n
```

# Графики рядов и функций

```python
n = n_sin(1,10**(-3))
y = []
x1 = np.linspace(-1,2,20)
for x in x1:
    y.append(math.sin(x))
x2 = np.linspace(0,1,10)
y1 = []
for x in x2:
    y1.append(sin_series(x,n))
fig, ax = plt.subplots(figsize=(7, 5), dpi=150)
ax.plot(x1,y, color = "limegreen", label = "sin(x)")
plt.legend()
plt.plot(x2,y1,color = 'darkgreen', label = "ряд, число слагаемих: " + str(n))
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("График 1, x ∈ [0,1]")
print("Число слагаемих ряда = ", n)
plt.minorticks_on()
plt.grid(which='major',
        color = 'grey', 
        linewidth = 0.7)
plt.grid(which='minor', 
        color = 'grey', 
        linestyle = ':')
plt.show()
```
